# Save pictures on Google Drive
To build our training data, we want to save some pictures. <br>
Here we will see howw to create directories directly from Google Colab and how to store pictures in it.

## Get an image
This part is not import. We just get ourselves an image from the web.

```
import numpy as np
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
import cv2

# get the image
url = "https://www.wemove.eu/sites/wemove.eu/files/baby_penguins_cropped.jpg"
response = requests.get(url)
img = np.array(Image.open(BytesIO(response.content)))

# print the image
plt.imshow(img)
plt.gca().axis('off')
plt.show()
```

## Mount the local environnement and create a directory

We can check the files and directories we can access using the 'os' library.

```
import os
def listdir(path="."):
  for f in sorted(os.listdir(path)):
    print(f)
```

```
listdir()
```

to extend this to our personal drive, we can mount our Drive. <br>
Those lines were on moodle. We now have a real use for them.

```
from google.colab import drive
drive.mount('/content/gdrive')
```

We check again to see the improvement.

```
listdir()
```

Our Drive is located into gdrive. Here we can check the content of our Colab Repository

```
path = "./gdrive/My Drive/Colab Notebooks/"
listdir(path)
```

We can even create a directory. If the directory already exists, this code will return an error, so don't be afraid to lose everything because of this. More info <a ref="https://www.geeksforgeeks.org/python-os-mkdir-method/">here</a><br>
We name our directory 'storage'

```
os.mkdir(path + "storage")
```

And we can even check its existence :

```
listdir(path)
```

## Save the picture
We now save the picture in our directory. Thanks Hugo Lavezac for this part <3.

```
cv2.imwrite(path + "storage/img.jpg", img)
```

We can also resize it.

```
resized = cv2.resize(img,(50,50))
cv2.imwrite(path + "storage/resized.jpg", resized)
```
